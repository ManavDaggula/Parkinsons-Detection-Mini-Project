```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# loading the data from csv file to a Pandas DataFrame
df = pd.read_csv('parkinsons.csv')
# printing the first 5 rows of the dataframe
df.head()
```

From here on we will be trying to do dataset cleaning and **transformations**.
Since the data is not having null values, we will proceed with **feature selection** and **dimensionality reduction**.

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
# print(df.columns)
col = [x for x in df.columns]
# print(col[5:9])
dropped_cols = df.drop(columns=col[5:9])
print(dropped_cols.columns)
col = [x for x in dropped_cols.columns]
# print(col[6:11])
dropped_cols = dropped_cols.drop(columns=col[6:11])
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(dropped_cols.corr(),annot=True)
plt.show()
```

```python
dropped_cols = dropped_cols.drop(columns=["NHR","MDVP:Shimmer"])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(dropped_cols.corr(),annot=True)
plt.show()
```

```python
dropped_cols.to_csv("transformed_data/without_Positive_Correlating_Attributes.csv")
```
